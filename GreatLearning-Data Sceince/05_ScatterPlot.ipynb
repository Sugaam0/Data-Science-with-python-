```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = [10, 20,30,40,50,60,70,80,90]
y1 = [6,8,9,2,5,4,7,6,8]
y2 = [4,8,1,3,8,9,4,5,3]
```

```python
plt.scatter(x,y1)
plt.show()
```

```python
plt.scatter(x,y1,marker="*", color='orange', s=200)
plt.show()
```

```python
plt.scatter(x,y1,marker="*", color='orange', s=200)
plt.scatter(x,y2,marker=".", color='yellow', s=300)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,y1,marker="*", color='orange', s=200)

plt.subplot(2,1,2)
plt.scatter(x,y2,marker=".", color='yellow', s=300)

plt.show()
```
