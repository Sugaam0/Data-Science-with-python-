### MatPlotlib

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(2,21)
x
```

```python
y = 3*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot()
plt.title("Line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x,y, color='r', linewidth=2, linestyle=":")
plt.show()
```

```python
x = np.arange(1,11)
x
```

```python
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1, color='g', linestyle=":")
plt.plot(x,y2,color='r', linestyle="-.")
plt.xlabel("Multiple plots")
plt.ylabel("Red and Green")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1, color='g', linestyle=":")

plt.subplot(2,1,2)
plt.plot(x,y2,color='r', linestyle="-.")

plt.show()
```
