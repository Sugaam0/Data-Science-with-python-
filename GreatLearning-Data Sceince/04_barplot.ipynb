### Bar plot

```python
from matplotlib import pyplot as plt
import numpy as np

student = {"Bob": 87, "Matt": 56, "Sam":27}
```

```python
Names = list(student.keys())
```

```python
Values = list(student.values())
```

```python
plt.bar(Names, Values)
plt.show()
```

```python
plt.barh(Names, Values, color="r")
plt.title("Bar plot")
plt.xlabel("Marks")
plt.ylabel("Names")
plt.show()
```
